```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Read in Homeownership file obtained from the Census Bureau https://www.census.gov/housing/hvs/data/ann17ind.html
# Used files Table 18. Homeownership Rates by Age of Householder for the Northeast [XLSX - 37 KB]
# Table 19. Homeownership Rates by Age of Householder for the Midwest [XLSX - 35 KB]
# Table 20. Homeownership Rates by Age of Householder for the South [XLSX - 30 KB]
# Table 21. Homeownership Rates by Age of Householder for the West [XLSX - 46 KB]

homeowner_file = "Resources/Homeownership_byAge_Region.csv"

homeowner_df = pd.read_csv(homeowner_file)

homeowner_df.head()
```

```python
#Create new dataframe with only 2015-2017 data and Age groups '25 to 29 years' and '30 to 34 years' to capture only 
#those homeowners considered millennials

new_homeowner = homeowner_df[['Region', 'Age group', '2015 ', '2016 ', '2017 ']].copy()
new_homeowner_df = new_homeowner.loc[(new_homeowner["Age group"] == "25 to 29 years") | (
    new_homeowner["Age group"] == "30 to 34 years"), :]

new_homeowner_df
```

```python
avocado_file = "Resources/Avocado_filtered.csv"

avocado_df = pd.read_csv(avocado_file)

avocado_df.head()
```

```python
new_avocado = avocado_df[['region', 'year', 'Total Volume']].copy()
new_avocado_df = new_avocado.groupby(['region','year']).sum()
new_avocado_df.reset_index(inplace=True)
new_avocado_df.head()
```
